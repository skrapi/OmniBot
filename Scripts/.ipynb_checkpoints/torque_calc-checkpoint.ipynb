# Torque Calculation

A script to calculate the size of the motors needed for the project

```python
import numpy as np
import pandas as pd
import sympy
import matplotlib.pyplot as plt
```

```python
class Motor:
    def __init__(self, mass_kg, nominal_voltage_v, speed_freerun_rpm, current_freerun_a, current_stall_a, torque_kgcm):
        self.mass_kg = mass_kg
        self.nominal_voltate_v = nominal_voltage_v
        self.speed_freerun_rpm = speed_freerun_rpm
        self.current_freerun_a = current_freerun_a
        self.current_stall_a = current_stall_a
        self.torque_kgcm = torque_kgcm
```

```python
motor = Motor(0.1, 12, 640, 0.8, 20, 0.14)
```

```python
torque = np.linspace(0, motor.torque_kgcm, 100)
speed = np.linspace(motor.speed_freerun_rpm, 0, 100)
```

```python
plt.plot(speed, torque)
plt.show()
```
